```python
from openpyxl import load_workbook
```

```python
workbook=load_workbook(filename='file1.xlsx')
```

```python
sheet=workbook.active
```

```python
col_val=['Vegeta','Sasuke','Zoro','Ishida','Levi']
```

```python
col_write=6
```

```python
for row_id,entry in enumerate(col_val,start=2):
    sheet.cell(row=row_id,column=col_write,value=entry)
```

```python
workbook.save('file1.xlsx')
```

```python
workbook.close()
```
