```python
from openpyxl import load_workbook
```

```python
workbook=load_workbook('file.xlsx')
```

```python
sheet=workbook.worksheets[0]
```

```python
row_val=['Eren','Yeager',22,'Titan','AOT']
```

```python
row_to_write=6
```

```python
for column_id,entry in enumerate(row_val,start=1):
    sheet.cell(row=row_to_write,column=column_id,value=entry)
```

```python
workbook.save('file.xlsx')
```

```python
workbook.close()
```
